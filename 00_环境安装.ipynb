# 00 环境安装
模型框架推荐使用Xinference。当然你也可以用一切能够封装为`langchain api`的大模型框架.

安装方法：
```shell
conda create xinference_env python=3.11
conda activate xinference_env
pip install "xinference[all]

启动xinference。
```shell
conda activate xinference_env
XINFERENCE_HOME=/home/你的工作目录 xinference-local --host 0.0.0.0 --port 9997
```
浏览器打开`http://0.0.0.0:9997`，启动语言模型和嵌入模型。首次启动需下载。

**注意**：下面的变量`server_url`和`model_uid`要换成你自己的。本项目的都要替换，后面将不再提示。

````python
'''
使用langchain框架调用Xinference LLM model。
'''
import os
from xinference.client import Client
from langchain_community.llms import Xinference

server_url=os.environ.get("SERVER_URL")
model_uid=os.environ.get("LLM_MODEL_UID")
prompt="如何安装python？"

llm = Xinference(
    server_url=server_url,
    model_uid=model_uid
)

response = llm(
    prompt=prompt,
    generate_config={"max_tokens":1024, "stream":True}
)
print(response)
````

````python
from langchain.chains import LLMChain
from langchain_core.prompts import PromptTemplate

template = "Where can we visit in the capital of {country}?"

prompt = PromptTemplate.from_template(template)

llm_chain = LLMChain(prompt=prompt, llm=llm)

generated = llm_chain.run(country="France")
print(generated)
````
